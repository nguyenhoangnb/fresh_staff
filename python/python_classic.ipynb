## 3. Cấu trúc điều khiển (if, for, while) ##

Lý thuyết:

- if - elif - else: điều kiện rẽ nhánh.

- for: lặp qua một dãy hoặc danh sách.

- while: lặp khi điều kiện đúng.

````python
x = 10

if x > 0:
    print("Số dương")
elif x == 0:
    print("Số 0")
else:
    print("Số âm")

for i in range(3):
    print("Lặp for:", i)

count = 0
while count < 3:
    print("Lặp while:", count)
    count += 1
````

## 5. Lập trình hướng đối tượng (OOP) ##

Lý thuyết:

- Lập trình hướng đối tượng (OOP) là phương pháp lập trình mô phỏng các thực thể trong thế giới thực như con người, xe hơi, sinh viên, v.v...

- OOP giúp chương trình::

    - Dễ mở rộng, bảo trì
    - Tái sử dụng mã (code reuse)
    - Tổ chức logic theo “đối tượng” (object) thay vì chỉ theo “hàm”

Tạo lớp (class) để mô tả đối tượng.

### 5.1 Tạo lớp (class), thuộc tính, và phương thức ###
- Class là khuôn mẫu (template) dùng để tạo ra đối tượng (object).
- Class định nghĩa các:
    - Thuộc tính (attributes) – dữ liệu mô tả đặc điểm của đối tượng
    - Phương thức (methods) – hành vi hoặc hành động của đối tượng
- Cấu trúc của một class
```python
class ClassName:
    # Hàm khởi tạo đối tượng (constructor)
    def __init__(self, tham_so1, tham_so2, ...):
        # Khai báo thuộc tính
        self.thuoc_tinh1 = tham_so1
        self.thuoc_tinh2 = tham_so2
    
    # Phương thức đối tượng (method)
    def method_name(self):
        # Hành động
        print("Đây là một phương thức trong class.")
```

#### Ví dụ: Lớp SinhVien ###

````python
class SinhVien:
    def __init__(self, ten, mssv):
        self.ten = ten
        self.mssv = mssv

    def hien_thi(self):
        print(f"Tên: {self.ten}, MSSV: {self.mssv}")

# Tạo đối tượng (object)
sv1 = SinhVien("Hoàng", "SV001")
sv1.hien_thi()
````

### 5.3. Attribute (Thuộc tính)

Khái niệm  
Attribute là biến dữ liệu lưu trữ thông tin của object hoặc class.

Phân loại

- Thuộc tính của đối tượng (Instance attribute)  
  - Khai báo trong `__init__()`  
  - Dùng riêng cho từng đối tượng

- Thuộc tính của lớp (Class attribute)  
  - Khai báo ngoài hàm `__init__()`  
  - Dùng chung cho mọi đối tượng

- Thuộc tính riêng tư (Private attribute)  
  - Đặt tiền tố `__` (hai gạch dưới) trước tên thuộc tính  
  - Không thể truy cập trực tiếp từ bên ngoài class (bị name-mangling)

Ví dụ minh họa:


Gợi ý ngắn
- Dùng instance attribute cho dữ liệu riêng từng object.  
- Dùng class attribute cho giá trị chung, hằng số.  
- Dùng private attribute khi cần ẩn trạng thái nội bộ; cung cấp method để truy cập/cập nhật an toàn.

````python
class Person:
    # Class attribute (dùng chung cho tất cả đối tượng)
    species = "Homo sapiens"

    def __init__(self, name, age):
        # Instance attributes (riêng cho từng object)
        self.name = name
        self.age = age
        # Private attribute
        self.__salary = 0  # ví dụ một thông tin nhạy cảm

    # Method để thay đổi giá trị private
    def set_salary(self, amount):
        self.__salary = amount

    # Method để lấy giá trị private
    def get_salary(self):
        return self.__salary

    # Method giới thiệu
    def introduce(self):
        print(f"Xin chào, tôi là {self.name}, {self.age} tuổi.")

# Sử dụng
p1 = Person("Hoàng", 25)
p2 = Person("Minh", 22)

# Truy cập class attribute
print(Person.species)

# Truy cập instance attributes
print(p1.name, p1.age)
print(p2.name, p2.age)

# Thay đổi và truy cập private attribute thông qua method
p1.set_salary(5000)
print("Salary:", p1.get_salary())

# Gọi method giới thiệu
p1.introduce()
p2.introduce()

# Thử truy cập private attribute trực tiếp sẽ lỗi/không khuyến cáo:
# print(p1.__salary)  # AttributeError
# Truy cập gián tiếp (name mangling)
print("Salary:", p1._Person__salary)  # 5000  (không khuyến cáo)
````

### 5.4 Methods (Phương thức)

Khái niệm  
Method là hàm định nghĩa trong class, dùng để xử lý dữ liệu hoặc thực hiện hành động của đối tượng.

Các loại method

| Loại method     | Cách định nghĩa                          | Mô tả                                      |
|-----------------|-----------------------------------------|--------------------------------------------|
| Instance Method | `def method(self, ...)`                 | Thao tác với object cụ thể (dùng `self`)    |
| Class Method    | `@classmethod` / `def method(cls, ...)` | Thao tác với class chung (dùng `cls`)       |
| Static Method   | `@staticmethod` / `def method(...)`      | Không dùng `self` hay `cls`, logic độc lập  |

Ví dụ ngắn:

`

Gợi ý ngắn
- Dùng instance method để thao tác trạng thái của object.  
- Dùng class method khi cần truy cập/điều phối trạng thái chung của class.  
- Dùng static method cho hàm tiện ích không phụ thuộc object/class.

````python
class ViDu:
    counter = 0  # class attribute

    def __init__(self, name):
        self.name = name
        ViDu.counter += 1

    def greet(self):               # instance method
        return f"Xin chào, {self.name}"

    @classmethod
    def total(cls):               # class method
        return f"Tổng đối tượng: {cls.counter}"

    @staticmethod
    def add(a, b):                # static method
        return a + b

# Sử dụng
o = ViDu("Hoàng")
print(o.greet())
print(ViDu.total())
print("Sum:", ViDu.add(2, 3))
````

# 6. Sử dụng thư viện threading để chạy đa luồng #


Lý thuyết  
`threading` là thư viện của Python cho phép chạy nhiều luồng (thread) trong cùng một tiến trình. Mỗi thread thực hiện một tác vụ riêng, hữu ích cho công việc I/O-bound hoặc khi cần xử lý song song các tác vụ nhẹ.

Khi dùng:
- Thích hợp cho I/O-bound (mạng, đọc/ghi file).  
- Với tác vụ CPU-bound, GIL giới hạn hiệu năng; cân nhắc `multiprocessing`.

Các bước cơ bản
1. Import thư viện: `import threading`  
2. Định nghĩa hàm (task) muốn chạy song song.  
3. Tạo đối tượng `threading.Thread` và gán hàm (target).  
4. Gọi `start()` để chạy thread.  
5. (Tùy chọn) Gọi `join()` để chờ thread kết thúc.




````python
import threading
import time

# Hàm chạy trên thread
def print_numbers():
    for i in range(1, 6):
        print(f"Số: {i}")
        time.sleep(1)

def print_letters():
    for letter in ['A', 'B', 'C', 'D', 'E']:
        print(f"Chữ: {letter}")
        time.sleep(1)

# Tạo thread
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

# Chạy thread
thread1.start()
thread2.start()

# Chờ các thread kết thúc
thread1.join()
thread2.join()
````

## 7. Cấu trúc switch-case trong Python

Lý thuyết  
Python không có câu lệnh `switch-case` truyền thống.  
Từ Python 3.10, ta có thể dùng `match-case` để thay thế — cho phép so khớp giá trị và mẫu (pattern matching) và thực thi nhánh tương ứng.

Cú pháp cơ bản:
```python
match value:
    case pattern1:
        # xử lý
    case pattern2:
        # xử lý
    case _:
        # default
```

Ví dụ đơn giản:
```python
def diem_to_xep_loai(diem):
    match diem:
        case 9 | 10:
            return "Xuất sắc"
        case 7 | 8:
            return "Giỏi"
        case 5 | 6:
            return "Trung bình"
        case _:
            return "Yếu/Kém"
```


Thay thế trên Python < 3.10  
Nếu dùng Python cũ hơn, có thể dùng `if-elif-else` hoặc dict mapping:
```python
# dict mapping
handlers = {
    "a": lambda: do_a(),
    "b": lambda: do_b(),
}
handlers.get(key, lambda: default())()
```

Gợi ý ngắn
- Dùng `match-case` khi cần so khớp mẫu phức tạp (tuple, dict, guard).  
- Với chỉ so sánh giá trị đơn giản, `dict` mapping hoặc `if-elif` vẫn gọn nhẹ.

## 8. Các hàm trong thư viện math

Lý thuyết  
Thư viện `math` cung cấp các hàm toán học cơ bản và nâng cao, dùng cho tính toán số học và lượng giác.

Một số hàm phổ biến

- `math.sqrt(x)` → căn bậc 2 của `x`  
- `math.pow(x, y)` → `x` mũ `y` (float)  
- `math.factorial(x)` → giai thừa của `x` (nguyên không âm)  
- `math.sin(x)`, `math.cos(x)`, `math.tan(x)` → các hàm lượng giác (đơn vị radian)  
- `math.radians(x)`, `math.degrees(x)` → chuyển đổi giữa độ và radian  
- `math.ceil(x)`, `math.floor(x)` → làm tròn lên / làm tròn xuống  
- `math.log(x[, base])`, `math.log10(x)`, `math.exp(x)` → logarit và hàm mũ  
- `math.pi`, `math.e` → hằng số pi và e


````python
import math

x = 9
y = 2

print(math.sqrt(x))       # 3.0
print(math.pow(x, y))     # 81.0
print(math.factorial(5))  # 120
print(math.sin(math.pi/2)) # 1.0
print(math.degrees(math.pi)) # 180.0
print(math.ceil(2.3))     # 3
print(math.floor(2.7))    # 2
````
